## Mean And Variance

## Mean

Mean is the measure of `central tendancy` in a dataset. The words mean and average can be used interchangably

Mean can be calculated with the following formular:

$\mu = \frac{1}{n}\Sigma_i x_i$

#### Calculations:

```python
import pandas as pd
```

```python
data = {
    'Numbers' : [16, 77, 78, 89, 10, 23, 24, 56, 78, 19, 51, 28, 36, 70, 47]
}
df = pd.DataFrame(data)
df.to_csv('dataset1.csv')
df
```

```python
df['Numbers'].mean()
```

## Variance($ \sigma^2 $)

In the same way that the mean is intended to describe the central tendency,
variance is intended to describe the `spread`. The variance of a set of values
is

Variance can be calculated using this formular:

$ \sigma^2 = \frac{1}{n}\Sigma_i (x_i - \mu)^2 $

#### Where:
$ \sigma $ = varaince\
$ \mu $ = mean

The term x i -μ is called the “deviation from the mean,” so variance is the
`mean squared deviation`, which is why it is denoted $ \sigma^2 $ . The square root of
variance, σ, is called the `standard deviation`.

> Note :
> #### Deviation:
the amount by which a single measurement differs from a fixed value such as the mean.

#### Calculations:

```python
df['devaition from mean'] = df['Numbers'] - df['Numbers'].mean()
df
```

```python
df['devaition from mean squared'] = df['devaition from mean']**2
df
```

```python
variance = df['devaition from mean squared'].mean()
variance
```

## Standard Deviation($ \sigma $)

Standard Deviation is the square root of the variance

**Formulars used with standard deviation**

1. When your data is the whole population: 

## $ \sigma = \sqrt[2]{\frac{1}{N}\Sigma_{i=1}^N(x - \mu)^2} $

2. When your data is a sample of the population:

## $ s = \sqrt[]{\frac{1}{N - 1}\Sigma_{i = 1}^N(x - \bar{x})^2} $

```python
import math
```

```python
standard_deviation = math.sqrt(variance)
standard_deviation
```
